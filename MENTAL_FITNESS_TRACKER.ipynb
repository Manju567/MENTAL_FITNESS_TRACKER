```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('/content/sample_data/Copy of prevalence-by-mental-and-substance-use-disorder.csv')
```

```python
data1.head()
```

```python
data2 = pd.read_csv('/content/sample_data/mental fitness tracker.csv')
```

```python
data2.head()
```

```python
data=pd.merge(data1,data2)
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop('Code',axis=1,inplace=True)
data.head()
```

```python
data.size,data.shape
```

```python
data.set_axis(['Country','Year','Schizophrenia','Bipolar disorder','Eating disorders ','Anxiety disorders','Drug use disorders','Depressive disorders','Alcohol use disorders','Mental disorders'],axis='columns',inplace=True)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot=True)
```

```python
sns.pairplot(data)
```

```python
mean=data['Mental disorders'].mean()
mean
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in data.columns:
    if data[i].dtype=='object':
        data[i]=le.fit_transform(data[i])
```

```python
data.shape
```

```python
data.head()
```

```python
x=data.drop('Mental disorders',axis=1)
y=data['Mental disorders']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score , classification_report, mean_squared_error, r2_score
regressor = RandomForestRegressor(n_estimators=100,
                                  random_state=0)
z = regressor.fit(X_train,Y_train)
y_train_pred = z.predict(X_train)
y_test_pred = z.predict(X_test)

print('MSE train: %.3f, test: %.3f' % (
        mean_squared_error(Y_train,y_train_pred),
        mean_squared_error(Y_test,y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2_score(Y_train,y_train_pred),
        r2_score(Y_test,y_test_pred)))
```
